<h3>Git Temelleri</h3>

<h4>Versiyon Kontrol Sistemleri (Version Control Systems - VCS)</h4>

<p><b>Tanım: </b>Versiyon kontrol sistemleri; bir proje kapsamında oluşturulan klasör ve dosyaların, zaman içerisinde geçirdiği değişiklikleri kaydeden, bu değişiklikleri takip eden ve yönetmeyi sağlayan sistemlerdir.</p>

<p>Diğer bir deyişle, versiyon kontrol sistemleri, bir projenin başlangıcından itibaren yapılan tüm değişiklikleri versiyonlar halinde saklar. Bu yüzden, kaydedilen her değişikliğe erişebilmenizi, gerekirse o noktaya dönebilmenizi, bir nevi geçmişe geri dönebilmenizi ve o noktadan sonra başka bir yol seyredebilmenizi, ve sizin uğraşmanıza gerek kalmayacak birçok kolaylığı size sağlamaktadır.</p>

<p>Yani, versiyon kontrol sistemleri, manuel olarak yönetildiğinde karmaşıklığa ve hataya açık olabilen proje yönetim süreçlerini otomatikleştirerek daha düzenli, güvenli ve sürdürebilir bir çalışma ortamı sunar.</p>

<h4>Versiyon Kontrol Sistemlerine Tarihsel Bir Bakış</h4>

<p>Birçok insan, yerel bilgisayarında bir klasör içerisine bir projenin farklı durumlarını farklı klasörlere kopyala yapıştır yöntemiyle saklamanın daha basit bir yöntem olduğunu düşünse de, bu yöntem hataya oldukça açıktır. </p> 

<p>Versiyon kontrol sistemlerinin sağladığı bir diğer kullanışlı özellik ise belli bir dosyayı otomatik olarak versiyonlamasından dolayı sizi kopyala yapıştır eyleminin zahmetinden kurtarmasıdır.</p>

<p>Bu yöntemin hataya açık yaklaşımına bir çözüm bulmak amacıyla yazılımcılar, uzun bir süre önce yerel versiyon kontrol sistemlerini geliştirmişlerdir.</p>

<h5>Yerel Versiyon Kontrol Sistemleri (Local VCS)</h5>

<p>Versiyon kontrol sistemlerinin en basit halidir ve proje dosyalarının farklı sürümleri aynı bilgisayar üzerinde saklanır ve yönetilir.</p>

<p>Basitliği, kurulum ve kullanım açısından bir avantaj gibi gözükse de, uzun vadede karmaşık projeler için yeterli değildir. Ayrıca, ekip çalışmasına da uygun değildir.</p>

<h5>Merkezileştirilmiş Versiyon Kontrol Sistemleri (Centralized VCS)</h5>

<p>İlerleyen zamanlarda yazılımcılar, ekip çalışması sorununa çözüm bulmak amacıyla merkezileştirilmiş versiyon kontrol sistemlerini geliştirmişlerdir.</p>

<p>Böylelikle, bir proje üzerinde kimin ne gibi değişiklik ya da değişikler yaptığının takibi yapılabilir olmuştur. Dolayısıyla, bir proje yöneticisi, bir ekip üyesinin ne yapabileceğini kontrol edebilir hale gelmiştir.</p>

<p>Ancak, bu tür bir versiyon kontrol sisteminin en büyük olumsuz yanı merkezi olmasından kaynaklanır. Eğer projenin başlatıldığı sunucu, büyük ihtimalle proje yöneticisinin bilgisayarı, çökmüşse hiç kimse projeye erişemez. Daha da kötüsü, projenin bulunduğu merkezi veri tabanı kullanılamaz hale gelirse ve düzgün bir yedekleme yapılmadıysa proje tamamen kaybedilebilir.</p>

<p>Aslında, projenin bulunduğu kaynağın kullanılamaz olma ihtimalinin bulunması riski hem yerel hem de merkezi versiyon sistemlerinin ortak sorunudur.</p>

<h5>Dağıtılmış Versiyon Kontrol Sistemleri (Distributed VCS)</h5>

<p>Yerel ve merkezileştirilmiş versiyon kontrol sistemlerinin yeterli olmadığı bu sorunu çözmek amacıyla dağıtılmış versiyon kontrol sistemleri geliştirilmiştir.</p>

<p>Projenin kaynağına erişimi olan istemciler, projeyi kendi bilgisayarlarına kopyaladıklarında yalnızca projenin en güncel hâline değil, aynı zamanda projenin tüm geçmişine de sahip olurlar.</p>

<p>Bu nedenle, projenin ana kaynağı ilerleyen zamanlarda çökse veya kullanılamaz hâle gelse bile, projeyi kopyalamış olan istemci depoları üzerinden projeye erişim mümkün olmaya devam eder. Ayrıca, her bir kopyalama aynı zamanda bir yedekleme niteliği taşır.</p>

<p>Bu tür bir sistem sayesinde, birçok farklı ekip birçok farklı şekilde aynı zamanda aynı proje üzerine çalışabilirler.</p>

<h4>Git</h4>

<p>Git, 2005 yılında Linus Torvalds tarafından geliştirilen, dağıtılmış bir versiyon kontrol sistemidir. Günümüzde hem açık kaynak hem de kurumsal projelerde en yaygın kullanılan versiyon kontrol sistemi haline gelmiştir.</p>

<h5>Temel İlk Kurulum Yapılandırması</h5>

<p>Bir önceki adımın son görselinde çalıştırılmış olan Git Bash'te her kullanıcının ilk kurulum sonrasında yapması gereken temel yapılandırmalar vardır.</p>

<p>Bu yapılandırmalar, Git'in kim olduğunuzu bilmesi için zorunludur. Yaptığınız ve yapılan tüm değişikliklerde bu bilgiler görünmektedir. Örneğin, bir projede birden fazla kişi çalışıyor ise kimin hangi değişikliği yaptığını bilmek için Git, bu bilgileri kullanmaktadır. Dolayısıyla, bu bilgiler oldukça önemlidir.</p>

Aşağıdaki git komutlarını bu yapılandırmaları gerçekleştirmek için sırasıyla Git Bash'e yazın:

<ul>
    <li>git config --global user.name "kullanıcı adınız"</li>
    <li>git config --global user.email "e-posta adresiniz"</li>
</ul>

<img src="screenshots/fig4.png" height="500" width="500"/>

<p>Yukarıdaki git komutlarını kendinize ait bilgiler ile doldurup yapılandırmayı tamamladıysanız aşağıdaki görselde olduğu gibi tanımlamış olduğunuz kullanıcı adı ve e-posta adresini <b>git config user.name</b> ve <b>git config user.email</b> komutlarını yazarak kontrol edebilirsiniz.</p>

<img src="screenshots/fig5.png" height="500" width="500"/>

<p>Her ne kadar yapılandırmalar sadece kullanıcı adı ve e-posta tanımlamakla sınırlı olmasa da, ilk defa git kullanan birisi için bu iki ayarı yapmak şimdilik yeterlidir. Ancak, merak ediyorsanız README.md dosyasında önerdiğim kaynak veya kaynaklardan daha detaylı bilgilere ulaşabilir, ya da resmi dokümantasyondan bilgi edinebilirsiniz.</p>

<h5>Git Temel Komutları</h5>

<p>Yukarıdaki kurulum aşamalarını tamamlandıktan sonra nihayet git temel komutları üzerine pratik yapabiliriz.</p>

<p>Git kullanmaya yeni başlayan bir kullanıcının sıkı sıkıya bağlı kalması gereken en çok kullanacağı komutlar şu şekilde listelenebilir:</p>

<ul>
    <li><b>git init:</b> bir git deposu oluşturmamızı sağlar.</li>
    <li><b>git status: </b> bir git deposunun o anki durumunu kontrol etmemizi sağlar.</li>
    <li><b>git log:</b> bir git deposunun değişiklik geçmişini görmemizi sağlar.</li>
    <li><b>git add:</b> bir git deposunda çalışma ortamında yapılan değişiklikleri Staging Area adındaki aşamaya eklemizi sağlar.</li>
    <li><b>git commit:</b> bir git deposundaki çalışmaları yerel depoya kesin olarak kaydetmemizi sağlar.</li>
    <li><b>git push:</b> bir git deposunda yerel depoya kaydedilen tüm değişiklikleri bağlanan uzak depoya göndermizi sağlar.</li>
    <li><b>git clone:</b> uzak bir sunucuda bulunan bir git deposunu klonlamamızı sağlar.</li>
</ul>

<p>Basit bir senaryo üzerinden tüm komutları, yeni başlayan birisinin pratiğe dökebileceği şekilde inceleyelim.</p>

<h5>Git'in Aşamaları / Git'in Yaşam Döngüsü</h5>

<p>Bir projede, bu senaryoda vao-ders-notlarim klasörü, git init ile bir git reposu başlattığınızda en başta var olan, ya da sonradan eklenecek her bir dosya ya da dosyaları içeren bir klasör, ilk başta git tarafından <b>Untracked (İzlenmeyen)</b> olarak adlandırılır.</p>

<p>Bunun anlamı, klasör içerisinde bulunan bu dosyaların henüz izlenme aşamasında olmamasıdır. Bu dosyalar, değiştirilmemiştir ya da izlenmemiştir.  Bu durumda ya git init ile klasör git deposuna dönüştürülmeden önce klasörün içinde zaten mevcutturlar ya da git init ile git deposuna dönüştürüldükten sonra projeye eklenmiştir.</p>

<p>Örneğin, <b>vao-ders-notlarim</b> klasörüne <b>TGSS 2024.sav</b> dosyasını ilk defa ekleyip ardından <b>git status</b> komutunu Git Bash'te çalıştırdığımda git tarafından bana bu dosyanın <b>Untracked</b> olduğunu söyleyen bir mesaj gösterilecektir.</p>

<p>TGSS 2024.sav dosyasını aşağıdaki gibi vao-ders-notlarim klasörüne ekleyin.</p>
<img src="screenshots/fig10.png" height="500" width="500"/>

<p>Sonrasında da Git Bash'te git status komutunu çalıştırın.</p>
<img src="screenshots/fig11.png" height="500" width="500"/>
<p>Görüldüğü üzere yukarıda bahsedilen mesaj, git tarafından gösterilecektir.</p>

<p>Bu dosyanın, git tarafından izlenmesini istersek <b>git add</b> komutunu kullanmamız gerekmektedir. Bu genel kullanımda iki yol vardır:</p>
<ul>
    <li>git add dosya_adi</li>
    <li>git add .</li>
</ul>

<p><b>git add dosya_adi</b>, belirli bir dosyanın git tarafından izlenmesini istediğimiz durumda kullanılırken, <b>git add .</b> o an <b>Untracked</b> olan tüm dosyaları izlemesini sağlamak için kullanılır.</p>

<p><b>3. Adım: </b>Proje klasörüme</b>, eklemiş olduğum TGSS 2024.sav dosyasının git tarafından izlenmesini sağlamak için <b>git add</b> komutunu, ardından da projenin son durumunu görmek için <b>git status</b> komutunu çalıştıralım.</p>

<p>İzleyeceğim adım sırası aşağıdaki gibidir:</p>
<ul>
    <li>git add TGSS 2024.sav ya da git add . komutunu çalıştır</li>
    <li>git status komutunu çalıştırarak çalışma dizininin son durumunu gör</li>
</ul>

<img src="screenshots/fig12.png" height="500" width="500"/>
<p>git add komutu ile çalıştırılan dosyalar, git terminolojisinde <b>Staging Area</b> olarak adlandırılan aşamaya eklenmektedir.</p>

<p><b>Staging Area</b>, bir dosyanın ya da dosyaların kaydedilmeden önce bulunduğu alan olarak düşünebilir. </p>

<p>Bu aşamaların daha iyi anlaşılması için gerçek hayata dayalı bir senaryo üzerinden anlatmakta fayda vardır.</p>

<p>Bir kitap yazdığımızı düşünelim. Kitap yazma süreci boyunca metin üzerinde sürekli değişiklikler yapıyoruz. Bu var olan bir metni değiştirmek de olabilir, tamamen yeni bir metin eklemek de olabilir. Bu çalışmalar, henüz kesinleşmemiş taslaklar olduğu için masamızın üzerindedir. Bu aşama, git terminolojisinde <b>Working Directory (Çalışma Dizini)</b> aşamasına karşılık gelir.</p>

<p>Bir süre çalıştıktan sonra, yazdığımız bazı metinlerin artık olması gerektiği noktaya geldiğini, ancak bu metinlerin kitaba kesin olarak girmesine karar vermediğimizi, bu yüzden de masadan alıp rafa kaldırdığımızı düşünelim. Raf, git terminolojisinde <b>Staging Area</b> dediğimiz aşamaya karşılık gelir. Bu aşamada bulunan metinler, kitaba girmeye adaydır, ancak son bir kontrol aşamasındadır. Dolayısıyla, <b>Staging Area</b> bir projedeki son kontrol aşamasıdır.</p>

<p><b>4. Adım:</b> TGSS 2024.sav dosyasının yerel git depomuza işlenmesi için <b>git commit</b> komutunu çalıştıralım ve ardından <b>git status</b> komutu ile projenin son durumunu kontrol edelim.</p>

<p>İzleyeceğim adım sırası aşağıdaki gibidir:</p>
<ul>
    <li>git commit -m "kaydedilen değişiklikleri temsil eden işleme mesajı" komutu ile <b>Staging Area</b> içerisinde bulunan dosyaları yerel depoya işle</li>
    <li>projenin son durumunu görmek için git status komutunu çalıştır</li>
</ul>

<img src="screenshots/fig13.png" height="500" width="500"/>

<p>Görüldüğü üzere, <b>git commit -m "TGSS 2024 adlı veri seti, projeye eklendi."</b> yazarak git add ile <b>Staging Area</b> aşamasına eklediğim dosyaları artık yerel depoya kaydettim. Buradaki -m yazacağımız mesaj için kullanılan bir flag idir. Flag ise bir komutun nasıl davranacağını yönlendiren ek bir seçenek olarak düşünebilir.</p>

<p>Ayrıca, dikkat edilmelidir ki, git commit komutunu kullanırken yazdığımız mesaj en azından yukarıda yazmış olduğum "TGSS 2024 adlı veri seti, projeye eklendi." kadar düzgün ve açıklayıcı olmalıdır. Çünkü, bu mesajlar, daha sonrasında siz tarafından, çalıştığınız ekip tarafından ve projenizi paylaştığınız diğer insanlar tarafından okunacak olan mesajlardır. Dolayısıyla, açıklayıcı olmak zorundadır.</p>

<p>Bu noktaya kadar git terminolojisinde bir dosyanın en temelde geçmesi gereken üç farklı aşamadan geçmesini sağladık:</p>
<ul>
    <li>Working Directory</li>
    <li>Staging Area</li>
    <li>Local Repository</li>
</ul>

<p><b>5. Adım: </b> <b>git log</b> komutuyla projemizde kesin kaydedilen değişikliklerin geçmişini, dönüm noktaları da diyebiliriz, görelim.</p>

<img src="screenshots/fig14.png" height="500" width="500"/>

<p>Görüldüğü üzere, bir önceki adımda git commit komutuyla yazmış olduğum mesaj, git log komutunu çalıştırdığımda git tarafından proje geçmişimde diğer meta veriler (veri hakkındaki veri) ile gösterilmektedir.</p>

<p>git log komutu, git commit komutu çalıştırıldığı an projeye yapılan her değişikliğin o anki tarihini (Date) ve değişikliği yapan kişi hakkındaki bilgileri (Author) göstermektedir. Dolayısıyla, bir ekibin bir üyesi başka bir zaman proje üzerinde bir değişiklik yapıp git commit komutuyla bu değişiklikliği kaydettiğinde, diğer ekip üyeleri de git log komutunu çalıştırarak bu değişikliği ve değişikliğin yapıldığı tarihi ve değişikliği yapan kişiyi görebilmektedir.</p>
