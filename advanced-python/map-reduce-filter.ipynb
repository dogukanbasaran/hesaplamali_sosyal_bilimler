<h3>Map, Reduce ve Filter Yerleşik Fonksiyonları</h3>

<h4>Map() Fonksiyonu</h4>

<p>Bir yinelebilirin her bir değerine bir fonksiyonu uygulayabilmemizi sağlar.</p>

<span>Örneğin:</span>

<p>Bir veri setinde yaş değişkeni için toplanan gözlemlerin bazılarının sayısal veri yerine metin biçiminde toplandığını varsayalım. Bu durumu düzeltmek için map fonksiyonundan yararlanalım.</p>

```python
age = [25, "24", 22, "23", "19", 21, "20"]
```

```python
def convert_string_to_int(value):
    if isinstance(value, str):
        return int(value)
    return value
```

```python
calculate_mean = round(sum(map(convert_string_to_int, age)) / len(age))
calculate_mean
```

<p>Ayrıca lambda fonksiyonunu map fonksiyonu ile kullanabiliriz.</p>
<span>Örneğin:</span>

```python
height = [177, 156, 170, 182, 191]
check_height = map(lambda height: "Ortalama üstü" if height > 175 else "Ortalama altı", height)
list(check_height)
```

<p>Map fonksiyonuna birden fazla yinebilir argüman geçilirse, fonksiyon da o kadar argüman kullanmak zorundadır.</p>
<span>Örneğin:</span>

```python
observations = {
    "height": [1.71, 1.60, 1.77],
    "weight": [76, 53, 82]
}
```

```python
def calculate_BMI (height, weight):
   return weight / (height ** 2)
```

```python
calculate_BMI = map(calculate_BMI, observations["height"], observations["weight"])
```

```python
list(calculate_BMI)
```

<h4>Reduce() Fonksiyonu</h4>

<p>Bir yinebilir içindeki değerleri tek bir değere indirgememizi sağlar. For döngüsünden daha okunaklı ve fonksiyoneldir.</p>

<span>Örneğin:</span>

```python
from functools import reduce
```

```python
def add(x, y):
    return x + y
```

```python
salaries = [55000, 32000, 180000]
```

```python
total = reduce(add, salaries)
```

```python
total
```

```python
print(f"average salary is {round(total/3)}₺.")
```

<p>Aynı sonucu lambda kullanarak da alabiliriz.</p>

```python
total = reduce(lambda x, y: x + y, salaries)
total
```

```python
print(f"average salary is {round(total/3)}₺.")
```

<h4>Filter() Fonksiyonu</h4>

<p>Bir yinelebilirde belirli bir şartı sağlayan değerleri döndürmemizi sağlar.</p>
<span>Örneğin:</span>

```python
grades = [50, 43, 75, 88, 23, 76, 78, 34]
```

```python
def check_grade(grade):
    if grade > 70:
        return grade
```

```python
grades_higher_than_70 = list(filter(check_grade, grades))
grades_higher_than_70
```

```python
f"Sınavı geçen kişi sayısı: {len(grades_higher_than_70)}"
```

<p>Aynı sonucu lambda kullanarak da alabiliriz.</p>

```python
grades_higher_than_75 = list(filter(lambda x: x > 75, grades))
f"Notu 75'ten yüksek olan kişi sayısı: {len(grades_higher_than_75)}"
```
