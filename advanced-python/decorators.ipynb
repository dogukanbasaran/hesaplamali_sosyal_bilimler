<h3>Decorators</h3>

<p>Bir fonksiyonu değiştirmeden genişletmemize olanak sağlayan fonksiyonlardır.</p>

<span>Örneğin:</span>



```python
def decorator_func(original_func):
    def wrapper_func(*args, **kwargs): 
        print(f"Bu mesajdan sonra {original_func.__name__} fonksiyonu çalışacaktır.")
        original_func(*args, **kwargs)
        
    return wrapper_func
```

```python
@decorator_func
def describe_yourself(name, university_name, department):
    print(f"Merhaba, benim adım {name}. {university_name} Üniversitesi'nde {department} bölümündeyim.")
```

```python
describe_yourself("Doğukan", "Marmara", "Sosyoloji")
```

<span>Örneğin:</span>

```python
def check_average_score(calculate_average_score):
    """ 
    Bu decorator fonksiyonun görevi:
        Hesaplanmış ortalama notun geçerli olup olmadığını kontrol eder ve sonucu ekrana yazdırır.
    NOT: 'Geçerli not' 70 üzeri ise, değilse 'Geçersiz not'
    """
    def wrapper(*args, **kwargs):
        average_score = calculate_average_score(*args, **kwargs)
        
        check_score_result  =  "Geçerli not" if average_score > 70 else "Geçersiz not"
        return check_score_result
    return wrapper
```

```python
@check_average_score
def calculate_average_score(score_1, score_2, score_3):
    """
    Bu fonksiyonun görevi:
        Girilen üç değerin (var sayalım ki bir ders için alınan notları temsil eden üç değer) ortalamasını hesaplar.
        Eğer bir değer 0'dan küçük veya 100'den büyük ise ValueError hatasını fırlatır.
    """
    scores = [score_1, score_2, score_3]

    exceeding_values = [score for score in scores if not 0 <= score <=100]
    if exceeding_values:
            raise ValueError(f"Sınırları aşan değerler girdiniz: {exceeding_values}")
        
    average_score = round((score_1 + score_2 + score_3) / 3, 2) 
    print(f"Ortalamanız: {average_score}")
    return average_score
```

```python
calculate_average_score(100, 75, 80)
```
