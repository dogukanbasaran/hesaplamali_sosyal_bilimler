<h3>Lambda Functions</h3>

<p>Bir kullanımlık, küçük ve anonim fonksiyonlardır. Daha temiz bir yazım sağlar ve map() gibi yüksek sıralı fonksiyonlar (Higher Order Functions) ile oldukça uyumludur.</p>
<p>İsteğe bağlı sayıda argüman alabilir, ancak sadece bir ifadeye sahip olabilir. Ayrıca klasik fonksiyonlardaki gibi açıklama veya dipnota sahip olamazlar.</p>

<h4>Örnekler</h4>

<h5>Örnek 1</h5>

```python
add = lambda x, y: x + y
```

```python
add(5, 10)
```

<h5>Örnek 2</h5>

```python
full_name = lambda first_name, last_name: f"{first_name} {last_name}"
```

```python
full_name("Doğukan", "Başaran")
```

<h5>Örnek 3</h5>

```python
grade_check = lambda grade: "Geçtiniz" if grade >= 70 else "Kaldınız"
```

```python
grade_check(50)
```

grade_check(100)

<h5>Örnek 4</h5>

```python
mean = lambda score_1, score_2, score_3: (score_1 + score_2 + score_3) / 3
```

```python
mean(77, 32, 90)
```

<h5>Örnek 5</h5>

```python
gpa_check = lambda mean: "Dersi geçtiniz" if mean >= 70 else "Dersten kaldınız"
```

```python
gpa_check(mean(100, 70, 70))
```
