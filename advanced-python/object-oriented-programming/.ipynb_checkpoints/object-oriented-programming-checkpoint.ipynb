<h3>Object Oriented Programming (Nesne Yönelimli Programlama)</h3>

<p><b>Tanım:</b> Nesne yönelimli programlama, temelinde, gerçek hayatta gördüğümüz şeyleri programlama dünyasında karşılığını oluşturmamıza olanak sağlayan bir yaklaşımdır.</p>

<p>Gerçek hayatta gördüğümüz şeylerden kasıt; insan, insanlar arasındaki ilişkiler, kurumlar, kurum ve insan arasındaki ilişkiler, bir araç, bir alan, bölüm, ders ya da görebileceğimiz herhangi bir şeydir.</p>

<p>Örneğin, sosyoloji öğrencisi olarak "Sosyolojiye Giriş" adlı bir dersi programlama dünyasında modellemek istersem bu dersin adı, bu derse o yıl kaydolan öğrenci sayısı, öğrencilerin listesi, dersi veren öğretim üyesinin adı gibi nitelikleri aktarabilirim. Ayrıca, burada bahsedilen öğrenci ve öğretim üyesi de modellenebilir.</p>

<p>Yanı sıra gerçek hayattaki bir şeye karşılık gelen bir modelin, Örneğin bir öğretim üyesinin, niteliklerinin olmasının yanında davranışlarını da yapılandırabiliriz. Örneğin, nesne yönelimli programlama, bir öğretim üyesinin dersi seçen öğrencilerin derse kaydını onaylaması eylemini yazılım nesnesine çevirmemize imkan tanır.</p>

<p>Bu bağlamda, nesne yönelimli programlama, gerçek hayattaki bir varlığın yazılım nesnesine çevrilmesi ve o nesnenin özelliklerinin ve davranışlarının bir araya toplanmasına imkan tanıyan bir programlama yöntemidir.</p>

<p>Nesne yönelimli programlama, Python programlama diline özgü değildir. JavaScript, C++, C#, Java, PHP gibi dillerde de bu yaklaşım vardır. </p>
    
<p>Bu yaklaşımın sunduğu genel ilkeler; encapsulation (kapsülleme), abstraction (soyutlama), inheritance (kalıtım), polymorphism (çok biçimlilik) idir.</p>

<ul>
    <li><b>Encapsulation (Kapsülleme)</b>, nesnenin istenmeyen detaylarını gizleyip sadece gerekli olan arayüzü    (interface) dışarı açma ilkesidir. Amaç, güvenlik, kontrol ve veri bütünlüğünü sağlamak ve modüle yapıyı teşvik etmektir.</li>
    <li><b>Abstraction (Soyutlama)</b>, gösterilmek istenmeyen detayları gizleyip önemli olan özellikleri öne çıkarma ilkesidir. Amaç, karmaşıklığı azaltmaktır.</li>
    <li><b>Inheritance (Kalıtım ya da Miras alma)</b>, bir sınıfın başka bir sınıfın niteliklerini ve yöntemlerini miras alma ilkesidir. Amaç, kod tekrarını azaltmak ve yeniden kullanabilirliği artırmaktır.</li>
    <li><b>Polymorphism (Çok biçimlilik)</b>, aynı arayüzün farklı şekillerde davranabilmesi ilkesidir. Bir metodu ezmemizi ve farklı parametrelerle birden fazla kez tanımlayabilmemizi sağlar. Amaç, esneklik ve genişletebilirlik sağlamaktır.</li>
</ul>
