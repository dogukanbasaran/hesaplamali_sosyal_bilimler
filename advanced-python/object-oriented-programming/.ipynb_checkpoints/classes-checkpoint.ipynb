<h3>Classes (Sınıflar)</h3>
<p>Sınıf, veri ve işlevselliği bir araya toplamamızı sağlayan, gerçek hayatta bir varlığın nasıl bir yazılım nesnesine modellenmesi gerektiğini gösteren plan ya da tasarımdır.</p>

<p>Örneğin, aşağıdaki sınıf gerçek hayatta bir dersin yazılımsal bir nesne olarak nasıl tanımlanabileceğini bize veren bir yönerge gibi düşünülebilir.</p>

```python
class Lecture:
    def __init__(self, name, quota, instructor):
        self.name = name
        self.quota = quota
        self.instructor = instructor
```

<p>Yukarıda bir dersin nasıl oluşturulması gerektiğini bize veren sınıftan bir "ders" nesnesi, diğer bir deyişle bu sınıfın bir örneğini, oluşturalım ve bu nesneyi oluştururken zorunlu tutulan "name", "quota" ve "instructor" özelliklerini aşağıdaki gibi tanımlayalım ve nesnemizi (ya da örneğimizi) oluşturalım.</p>

```python
lecture_1 = Lecture("Programlamaya giriş", 50, "Doğukan Başaran")
```

<p>Bu örneğin üzerinden oluştururken verdiğimiz özellikleri aşağıdaki gibi çağırabiliriz.</p>

```python
lecture_1.name
```

```python
lecture_1.instructor
```

<p>Aynı zamanda, bir sınıftan istediğimiz kadar nesne, ya da o sınıfın örneğini, oluşturabiliriz.</p>

```python
lecture_2 = Lecture("Sosyolojiye Giriş", 100, "Prof. Dr. Mehmet Fatih Aysan")
```

```python
lecture_2.name
```

```python
lecture_2.quota
```

```python
lecture_2.instructor
```

<h4>Nitelik/Özellik (Attribute)</h4>

<p>Belirttiğimiz üzere name (dersin adı), quota (derse kayıt olabilecek öğrenci sayısının sınırı) ve instructor (dersi veren öğretim üyesi), <b>Lecture</b> sınıfından türetilen bir nesnenin zorunlu olarak en baştan sahip olması gereken özellikleri/nitelikleridir.</p>

<p>Örneğin, Lecture sınıfında __init__ yapıcı (constructor) fonksiyonu içerisinde bulunan nitelikleri yeni bir "ders" nesnesi oluştururken şu şekilde belirleyelim:</p>

```python
lecture_3 = Lecture("Yapay Zekaya Giriş", 50, "Ekinsu Ülgen")
```

<p>Yani burada yaptığımız şey aslında <b>Lecture</b> sınıfında belirtilen name özelliğini <b>Yapay Zekaya Giriş</b>, quota özelliğini <b>50</b> ve instructor özelliğini <b>Ekinsu Ülgen</b> olarak belirleyerek Lecture sınıfından bir nesne oluşturmak oldu.</p>

<p>Sınıflar, bize gerçek hayattaki bir varlığın sadece özelliklerini değil, aynı zamanda davranışlarını da modelleyebilmemizi sağlar. Bunu yaparken de metotları kullanırız.</p>

<h4>Metot (Method)</h4>

<p>Metot ise belirli bir sınıfa ait olan fonksiyonlardır ve o sınıfa ait özellikleri/nitelikleri kullanarak gerçek hayattaki bir varlığın bir eylemini temsil eder.</p>

<span>Örneğin:</span>

```python
class Student:
    def __init__(self, full_name, enrolled_lectures):
        self.full_name = full_name
        self.enrolled_lectures = enrolled_lectures

    def display_enrolled_lectures(self):
        print(f"Toplamda {len(self.enrolled_lectures)} derse kayıtlısınız ve kayıtlı olduğunuz dersler şunlardır:")

        counter = 1
        for lecture in self.enrolled_lectures:
            print(f"{counter}. {lecture.name}")
            counter += 1
```

```python
student_1 = Student("Doğukan Başaran", [lecture_1, lecture_2])
```

```python
student_1.display_enrolled_lectures()
```

<p>Görüldüğü üzere student_1 değişkenine atanmış bir nesne üzerinden çağırdığımız <b>display_enrolled_lectures</b> metodu bize şu bilgileri sağlamakta:</p>

<ul>
    <li>Öğrencinin kaydolduğu ders sayısı</li>
    <li>Öğrencinin kaydolduğu derslerin adları</li>
</ul>

<p>Yukarıda tanımlanmış lecture_1, lecture_2 ve lecture_3 dersleriyle yeni bir öğrenci nesnesi oluşturalım.</p>

```python
student_2 = Student("Yavuzhan Melikşah Şenol", [lecture_1, lecture_2, lecture_3])
```

```python
student_2.display_enrolled_lectures()
```

<p>student_1 değişkenine atanmış olan öğrenci nesnesi iki derse kayıtlıyken, student_2 değişkenine atanmış olan öğrenci nesnesi ise üç derse kayıtlıdır.</p>

<h4>__init__ Yapıcı Metodu ve self Parametresi</h4>

<p>Yukarıdaki örneklerden de anlaşılacağı üzere <b>__init__</b>, bir sınıftan türetilen bir nesnenin oluşturulduğu zaman otomatik olarak çalışan özel bir metottur.</p>

<p>Her nesne için ayrı ayrı çalışır ve nesnenin başlangıç değerlerini (niteliklerini) belirtmek için kullanılır.</p>

<p>Örneğin, aşağıdaki Instructor sınıfında __init__ yapıcı metodu, bu sınıftan türetilecek her nesne için başlangıçta name ve university ve department özelliklerini belirlememizi gerektirir.</p>

```python
class Instructor:
    def __init__(self, name, university, department):
        self.name = name
        self.university = university
        self.department = department
```

```python
instructor_1 = Instructor("Doç. Dr. Melih Çoban", "Marmara Üniversitesi", "Sosyoloji")
```

```python
instructor_1.name
```

```python
instructor_1.university
```

```python
instructor_1.department
```

<p><b>self</b> parametresi ise, oluşturulan bir nesnenin kendisini temsil eder ve sınıf içinde belirtilen niteliklere ve metotlara erişmemize olanak tanır.</p>

<p>Örneğin, instructor_1 nesnesi için Instructor sınıfındaki self.name alanını instructor_name olarak düşünebiliriz ve bunun karşılığı instructor_1 için "Doç. Dr. Melih Çoban" idir.</p>

<span>Örneğin:</span>

```python
instructor_2 = Instructor("Dr. Öğr. Üyesi Lütfiye Zeynep Beşpınar", "Marmara Üniversitesi", "Sosyoloji")
```

```python
instructor_2.name # Bu nesne için self.name, "Dr. Öğr. Üyesi Lütfiye Zeynep Beşpınar" idir.
```
